<h5>Modulo Operator % </h5>
<p>This operator gives the remainder when first number is divided by second number. i.e a%b=c, when b*d+c=a, and d = a//b </p>

<h5>Nested if and elif statements</h5>
<p>if-elif-else written inside a parent if-elif-else is known as nested statement</p>
<pre>
if condition_one: 
    if condition_two:
        statement to execute if condition_one and condition_two both are true
    else:
        statement to execute if condition_one is true and condition_two is false
else:
    if condition_three:
        statement to execute if condition_one is false and condition_three is true.
    else:
        statement to execute if condition_one is false and condition_three is false.
</pre>
