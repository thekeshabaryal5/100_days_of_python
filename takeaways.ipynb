### loops in python

#### for loop
- syntax: 
```python
for x in iterable:
    do_something..
```
iterable means a collection where each items can be accessed separately
- we use for loop when we know exactly how many times the loop must run.
- for e.g:
```python
for i in range(10):
    print(i)
```
this prints all integer from 0 till 10(exclusive)

#### while loop
- syntax:
```python
while condition:
    do_something
```
- this loops runs as long as the condition is true. Once the condition gets false, loop terminates.
- while loop is useful when we don't know in advance, how many time we should run the loop.
- It is useful in the cases where the condition depend upon user input, or the conditions where the loop numbers are unknown.

#### List methods:

##### indexing
- list items can be accessed via index (index starts from 0 and goes till length_of_list(exclusive))
- e.g:
```python
numbers = [1,2,3,4,5]
print(numbers[0]) # 1
print(numbers[3]) # 4
print(numbers[5]) # gives error, as index 5 is not in the numbers
```

#### slicing
- slicing refers to taking a portion of a list (can be one element or more)
- syntax: **list1[a:b]**
- return a new list having an item from index a till index b(exclusive) of list1
- if first index, i.e. a is omitted list1[:b], it returns a new list containing all the items from index 0 till index b(exclusive)
- if second index, i.e. b is omitted list[a:], it returns a new list containing all the items from index a till last element(inclusive).
```python
numbers = [1,2,3,4,5]
sliced_number = numbers[2:4]
```
- here a new list is created, which contains items from index 2 till index 4(exclusive) of numbers. So sliced_number is a new list having 3 and 4 as its members. i.e sliced_number =[3,4]

#### concatenation
- list also supports concatenation. Concatenation means adding one list to another.
- syntax: list1 + list2
- returns new list having all the elements of list1 and list2
- e.g:

````python
list1 = [1,2,3,4]
list2 = [4,5,6,7]
list3 = list1+list2
print(list3)
````

#### assignment
- unlike string, list is mutable data type, i.e. it supports re-assignment of item.
- re-assignment is done by selecting the item via index and assigning a new value to it.
-e.g:

````python
numbers = [1,2,3,4,5]
print("Changing the item 3 to 33")
numbers[2] = 33
print(numbers)
````

#### append
- this function is used for adding a new element/item/member at the end of the list.
- syntax: list1.append(item)
- e.g:

````python
list1 = [1,2,3,4,5]
print(f"Before appending {list1}")
list1.append(10)
print(f"After appending {list1}")
````

#### copy of list
- we can make a copy of list by using copy() method and slicing method
- copy() method is a in-built method and create a copy of a given list and assign it to new list.
- syntax: **new_list = list1.copy()**

````python
original_list = [1,2,3,4,5]
list_copy = original_list.copy()
print(list_copy) # prints the copy of original list
list_copy.append(14) # appending a item to copy list
print(list_copy) # appending item to copy of list won't affect the original list
print(original_list)
````

- another way to make a copy of list is using slicing.
- syntax: **new_list = original_list[0:len(original_list)]**
- here we are slicing original list from beginning to end, which can be more simplified as **new_list = original_list[:]**

````python
original_list = [1,2,3,4,5]
list_copy = original_list[:]
print(list_copy) # prints the copy of original list
list_copy.append(14) # appending a item to copy list
print(list_copy) # appending item to copy of list won't affect the original list
print(original_list)
````

- why we didn't simply assign the original list to new variable ?
- **list_copy = original_list** this looks simple, but there is a catch.
- simply assigning a original list to a variable will cause an error, as both list points to a same memory location. i.e. both are same list, having different name.

````python
original_list = [1,2,3,4,5]
list_copy = original_list
print(id(original_list))
print(id(list_copy)) # same id as original_list

list_copy.append(14) # appending a item to copy list
print(list_copy) 
print(original_list) # here original list will be same as original and copy list are pointing to same object.
````

### finding the length of a list. len() function
- len() function returns a length of a list passed as an argument to it.
- returns an integer value.
- syntax:
```python
length_of_list = len(list_name) # returns the length of list
```

````python
list1 = [1,2,3,4,5]
length_of_list = len(list1) # returns 5
print(length_of_list) # prints 5
````

### inserting an item
- we can insert a new item at any position of list by using **insert(a,b)** function, where a is the index and b is the value.
- length of list is increased by 1 every times we insert a new item.
- return None and original list is affected

````python
list1 = [1,2,3,4,5]
list1.insert(3,59) # inserts 59 at index 3
print(list1)
````

### other methods
- **list.remove(x)**    removes the item x from list
- **list.pop([i])**     removes the item at index i and returns it. If i is omitted, last element is removed
- **list.clear()**      remove all the items from the list.
- **list.count(x)**     returns the occurrence of x inside the list
- **list.reverse()**    reverse the list
